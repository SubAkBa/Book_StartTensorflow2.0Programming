```
# 예제 3.31 : 간단한 꺾은선 그래프 그리기
import tensorflow as tf
import matplotlib.pyplot as plt
x = range(20)
y = tf.random.normal([20], 0, 1)
plt.plot(x, y)
plt.show()
```

```
# 예제 3.32 : 간단한 점 그래프 그리기
import matplotlib.pyplot as plt

x = range(20)
y = tf.random.normal([20], 0, 1)
plt.plot(x, y, 'bo')
plt.show()

# 선 : 'b-'
# 점선 : 'b--'
# 색깔 : 빨간색 (r), 노란색 (y), 초록색 (g), 검은색 (k)
```
